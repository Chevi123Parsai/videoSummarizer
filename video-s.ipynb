```python

import whisper_timestamped as whisper

audio = whisper.load_audio('/kaggle/input/dbms-dataset/Introduction to Relational Model_2.mp4')
model = whisper.load_model("base")

result = whisper.transcribe(model, audio, language="en")

import json
# print(json.dumps(result, indent = 2, ensure_ascii = False))
```

```python

maintext = json.dumps(result['text'] , indent = 2, ensure_ascii = False)

print(maintext)
```

```python
times = []
for i, segment in enumerate(result['segments']):
    start, end = segment['start'], segment['end']
    print(i)
    print(f"00:00:{str(int(start)).replace('.', ',')} --> 00:00:{str(int(end)).replace('.', ',')}")
    times.append(int(start))
    print(segment['text'].strip())
    print()
```

```python
import cv2
import os
from datetime import datetime

#path to save images
path = r'/kaggle/working/'
os.chdir(path)

i = 0

video = cv2.VideoCapture('/kaggle/input/dbms-dataset/Introduction to Relational Model_2.mp4')

fps = video.get(cv2.CAP_PROP_FPS)
print('frames per second =',fps)
```

```python
for i in times:
        
    frame_id = int(fps*( i))
#     print('frame id =',frame_id)
    video.set(cv2.CAP_PROP_POS_FRAMES, frame_id)
    ret, frame = video.read()
    
    cv2.imwrite(str(i)+ '.png', frame)
```
